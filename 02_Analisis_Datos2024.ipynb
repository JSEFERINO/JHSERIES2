```python
import re
import prince
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (15,6)
warnings.filterwarnings("ignore")
url_base = "https://raw.githubusercontent.com/lacamposm/Metodos_Estadisticos/main/data/"
```

```python
# @title Importar datos
datos = pd.read_csv('/content/datos_estudiantes.csv', delimiter=',', encoding='latin-1')
datos.head(5)
```

```python
# Verificar el tipo de datos de cada columna
datos.dtypes
```

```python
# Definir la función para categorizar los valores
def categorizar_defintiva2(valor):
    if 0 <= valor < 2.5:
        return 'Bajo'
    elif 2.5 <= valor < 3.5:
        return 'Medio'
    elif 3.5 <= valor <= 5:
        return 'Alto'
    else:
        return 'Fuera de rango'  # Para valores que no estén en los rangos especificados

# Aplicar la función a la columna 'DEFINITIVA2' y crear una nueva columna 'Categoría'
datos['Categoria'] = datos['DEFINITIVA2'].apply(categorizar_defintiva2)

# Mostrar los primeros 5 registros para verificar
print(datos.head())
```

```python
df2 = datos[['ESTRATO']].join(datos.iloc[:, 16:])

# Mostrar las primeras filas del nuevo DataFrame
print(df2.head())
```

```python
#remover columnas
df2 = df2.drop(['Categoria_DEFINITIVA2', 'Categoría'], axis='columns')
df2 = df2.astype("category")
df2.info()
```

```python
# Variables dummies.
X = pd.get_dummies(df2, dtype=int)
X.head()
```

```python
# Realizamos CA en X que es get_dummies del df2
ca = prince.CA(n_components=7, random_state=42).fit(X)
ca
```

Se extrae la información del análisis

```python
ca.eigenvalues_summary
```

3. Para interpretar los datos solo se tendrán en cuenta los 2 primeros factores.



```python
# Tabla con información sobre los 2 primeros factores.
tmp = ca.column_coordinates(X)[[0,1]].rename(columns={0: "factor_1", 1: "factor_2"})
tmp1 = ca.column_contributions_
tmp1 = pd.DataFrame(tmp1).iloc[:, [0,1]]
tmp1.columns = ["Dim1", "Dim2"]
tmp1.index = tmp.index

tmp.merge(tmp1, left_index=True, right_index=True)
```

# **1.1. Introducción de series de tiempo**
+ Una serie temporal es simplemente un conjunto de puntos de datos ordenados en el tiempo.
+ Además, los datos suelen estar igualmente espaciados en el tiempo, lo que significa que intervalos iguales separan cada punto de datos.
+ En términos más simples, los datos se pueden registrar cada hora o cada minuto, o se pueden promediar cada mes o año.
+ Algunos ejemplos típicos de series temporales incluyen el valor de cierre de una acción en particular, el consumo de electricidad de un hogar o la temperatura exterior.

## **Descomposición de series temporales**
+ La descomposición de series de tiempo es un proceso mediante el cual separamos una serie de tiempo en sus componentes: tendencia, estacionalidad y residuos.

+ **La tendencia** representa los cambios lentos en una serie de tiempo. Se encarga de hacer que las series aumenten o disminuyan gradualmente con el tiempo.

+ **El componente de estacionalidad** representa el patrón estacional de la serie. Los ciclos ocurren repetidamente durante un período de tiempo fijo.

+ **Los residuales** representan el comportamiento que no puede explicarse por los componentes de tendencia y estacionalidad. Corresponden a errores aleatorios, también denominados ruido blanco.

+ Ya podemos ver intuitivamente cómo cada componente afecta nuestro trabajo al realizar pronósticos.
+ Si una serie temporal expone una determinada tendencia, esperaremos que continúe en el futuro.
+ De manera similar, si observamos un fuerte efecto estacional, es probable que esto continúe y nuestras previsiones deben reflejarlo.
+ Más adelante en el libro, verá cómo tener en cuenta estos componentes e incluirlos en sus modelos para realizar pronósticos más complejos.

# **1.3. En qué se diferencia el pronóstico de series de tiempo de otras tareas de regresión**
+ Probablemente haya encontrado tareas de regresión en las que debe predecir algún objetivo continuo dado un determinado conjunto de características.
+ A primera vista, el pronóstico de series de tiempo parece un problema de regresión típico: *tenemos algunos datos históricos y deseamos construir una expresión matemática que exprese los valores futuros en función de los valores pasados. Sin embargo, existen algunas diferencias clave entre el pronóstico de series de tiempo y la regresión para escenarios independientes del tiempo que merecen ser abordadas antes de analizar nuestra primera técnica de pronóstico*.

## **1.3.1. Las series temporales tienen un orden**
+ El primer concepto a tener en cuenta es que las series temporales tienen un orden y no podemos cambiar ese orden al modelar.
+ En el pronóstico de series de tiempo, expresamos los valores futuros en función de los valores pasados.
+ Por ello, debemos mantener los datos en orden, para no vulnerar esta relación.

+ Además, tiene sentido mantener los datos en orden porque su modelo sólo puede utilizar información del pasado hasta el presente; no sabrá qué se observará en el futuro.
+ Recuerde su viaje de campamento. Si desea predecir la temperatura del martes, no es posible utilizar la información del miércoles, ya que, desde el punto de vista del modelo, es futura.
+ Sólo podrás utilizar los datos del lunes y antes. Es por eso que el orden de los datos debe permanecer igual durante todo el proceso de modelado.

+ Otras tareas de regresión en el aprendizaje automático a menudo no tienen un orden.
+ Por ejemplo, *si tiene la tarea de predecir los ingresos en función de la inversión publicitaria, no importa cuándo se gastó una determinada cantidad en anuncios.*
+ En su lugar, simplemente desea relacionar la cantidad de inversión publicitaria con los ingresos.
+ De hecho, incluso podrías mezclar aleatoriamente los datos para que tu modelo sea más sólido.
+ Aquí la tarea de regresión consiste simplemente en derivar una función tal que, dada una cantidad de inversión publicitaria, se devuelva una estimación de los ingresos.

+ Por otro lado, las series temporales están indexadas por tiempo, y ese orden debe mantenerse.
+ De lo contrario, estaría entrenando su modelo con información futura que no tendría en el momento de la predicción.
+ Esto se llama **sesgo de anticipación** en términos más formales.
+ Por lo tanto, el modelo resultante no sería fiable y muy probablemente tendría un rendimiento deficiente cuando se realicen análisis futuros.

## **1.3.2. Las series temporales a veces no tienen características**
+ Es posible pronosticar series de tiempo sin el uso de otras características que la propia serie de tiempo.

+ Como científicos de datos, estamos acostumbrados a tener conjuntos de datos con muchas columnas, cada una de las cuales representa un predictor potencial para nuestro objetivo.
+ **Por ejemplo, considere la tarea de predecir los ingresos en función de la inversión publicitaria, donde los ingresos son la variable objetivo. Como características, podríamos tener la cantidad gastada en anuncios de Google, anuncios de Facebook y anuncios de televisión. Utilizando estas tres características, construiríamos un modelo de regresión para estimar los ingresos.**

+ Sin embargo, con las series temporales, es bastante común recibir un conjunto de datos simple con una columna de tiempo y un valor en ese momento.
+ Sin ninguna otra característica, debemos aprender formas de utilizar valores pasados ​​de la serie temporal para pronosticar valores futuros.
+ Aquí es cuando entran en juego el **modelo de media móvil (capítulo 4)** o **el modelo autorregresivo (capítulo 5)**, ya que son formas de expresar valores futuros en función de valores pasados.
+ Estos modelos son fundamentales para los modelos más complejos que luego le permiten considerar patrones estacionales y tendencias en series temporales.
